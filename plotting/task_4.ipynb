```python
# Imports 
import numpy as np 
import matplotlib.pyplot as plt 
```

```python
data = np.genfromtxt("../data/task4.csv", delimiter=",")[:, :-1]
betas = [0.6, 0.7, 0.8, 0.9, 1]
```

```python
for values, beta in zip(data, betas): 
    plt.plot(values, label=r"$\beta$ = {0}".format(beta))
plt.legend()
plt.ylabel(r"$\alpha$")
plt.xlabel("Optimization steps ($p$)")
plt.show()
```
