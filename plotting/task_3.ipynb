```python
# Imports
import numpy as np 
import matplotlib.pyplot as plt 
plt.rcParams['figure.dpi'] = 200
```

```python
data = np.genfromtxt("../data/task3.csv", delimiter=",")
```

```python
plt.errorbar(data[:, 0], data[:, 1], yerr=np.abs(data[:, 2]), fmt='o', ecolor="green", label=r"$\sigma$")
plt.xlabel(r"$\alpha$")
plt.ylabel("$E_L$  [a.u.]")
plt.legend()
plt.show()
```
